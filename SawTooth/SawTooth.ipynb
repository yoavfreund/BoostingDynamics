```python
def phi(x,s,m):
    out=(2*x+(s*(2**(-m))*sin((2**m)*x))) % 1.0
    return out
```

```python
x=arange(0,1,0.01)
m=2
s=0.01

y=phi(x,s,m)
plot(x,y);
```

```python
def iter(z=0.1):
    L=[z]
    for i in range(100):
        z=phi(z,s,m)
        L.append(z)
    return L
```

```python
L1=iter(0.1)
L2=iter(0.1001)
plot(L1);
plot(L2);
```

```python
def gen_envelope(x=0.1,epsilon0=0.00001,n=100):
    _range=arange(x-epsilon0,x+epsilon0,epsilon0/n)
    LL=[]
    for z in _range:
        LL.append(iter(z))
    M=array(LL)
    _min=np.min(M,axis=0)
    _max=np.max(M,axis=0)
    return _min,_max
```

### Measuring the length of stability

for each value of x, add a small noise $\epsilon_0=0.00001$ and find the 
number of iterations where the deviation is smaller than $\epsilon_1 = 0.1$.

This time should depend only weakly on the choice of $\epsilon_0,\epsilon_1$

```python
_min,_max=gen_envelope()
plot(_max-_min);
```

```python
epsilon1=0.5
I=[]
X=arange(0,1,0.01)
for x in X:
    _min,_max=gen_envelope(x)
    gap=_max-_min
    I.append(nonzero(gap>epsilon1)[0][0])
```
